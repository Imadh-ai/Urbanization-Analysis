```python
import os
import re
import glob
import numpy as np
import rasterio
from rasterio.mask import mask
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from datetime import datetime
from pathlib import Path

LANDSAT_DIR = Path("/Users/imadh/Downloads/My version/Rearrangedats")
AOI_PATH = Path("/Users/imadh/Downloads/My version/AOI/Kurunegala_District_AOI.geojson")
OUTPUT_DIR = Path("/Users/imadh/Downloads/My version/URB_outputs")

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Years for your project
YEARS = range(2015, 2021)

# Landsat Collection 2 Level 2 bands
BANDS = {
    "blue": "SR_B2.TIF",
    "green": "SR_B3.TIF",
    "red": "SR_B4.TIF",
    "nir": "SR_B5.TIF",
    "swir": "SR_B6.TIF",
}

# =====================================================
# LOAD AOI
# =====================================================

aoi = gpd.read_file(AOI_PATH)
print(f"AOI loaded: {len(aoi)} polygon(s)")
print(f"AOI CRS: {aoi.crs}")

# =====================================================
# FUNCTIONS
# =====================================================

def extract_date_from_name(name):
    """
    Extract date like 20150203 from Landsat filename/folder name.
    """
    match = re.search(r"(20\d{6})", name)
    if match:
        date_str = match.group(1)
        return datetime.strptime(date_str, "%Y%m%d")
    return None


def clip_to_aoi(raster_path, aoi_geom):
    """
    Clip one raster band to AOI.
    """
    with rasterio.open(raster_path) as src:
        aoi_reprojected = aoi_geom.to_crs(src.crs)
        out_img, out_transform = mask(src, aoi_reprojected.geometry, crop=True)
        out_meta = src.meta.copy()

        out_meta.update({
            "driver": "GTiff",
            "height": out_img.shape[1],
            "width": out_img.shape[2],
            "transform": out_transform
        })

    return out_img[0].astype("float32"), out_meta


def find_landsat_scenes():
    """
    Find all Landsat scenes inside arranged dataset folder.
    Works with structure:
    Landsat_Arranged/
       2015/
          20150203/
             LC08_..._SR_B4.TIF
             LC08_..._SR_B5.TIF
    """
    scenes = []

    for year in YEARS:
        year_folder = LANDSAT_DIR / str(year)

        if not year_folder.exists():
            print(f"Year folder not found: {year_folder}")
            continue

        red_files = list(year_folder.rglob("*SR_B4.TIF"))

        for red_file in red_files:
            date = extract_date_from_name(red_file.name)

            if date is None:
                date = extract_date_from_name(str(red_file.parent))

            if date is None:
                print(f"Could not find date for: {red_file}")
                continue

            scene_id = red_file.name.replace("_SR_B4.TIF", "")

            scene_info = {
                "scene_id": scene_id,
                "date": date,
                "year": date.year,
                "month": date.month,
                "folder": red_file.parent
            }

            scenes.append(scene_info)

    return sorted(scenes, key=lambda x: x["date"])


def find_scene_files(scene):
    """
    Find all required band files for one scene.
    """
    scene_files = {}
    missing_bands = []

    for band_name, suffix in BANDS.items():
        file_path = scene["folder"] / f"{scene['scene_id']}_{suffix}"

        if file_path.exists():
            scene_files[band_name] = file_path
        else:
            missing_bands.append(band_name)

    return scene_files, missing_bands


def calc_indices(red, nir, swir):
    """
    Calculate NDVI and NDBI using Landsat Collection 2 scale factor.
    """
    red = red * 0.0000275 - 0.2
    nir = nir * 0.0000275 - 0.2
    swir = swir * 0.0000275 - 0.2

    ndvi = (nir - red) / (nir + red + 1e-10)
    ndbi = (swir - nir) / (swir + nir + 1e-10)

    ndvi = np.where(np.isfinite(ndvi), ndvi, np.nan)
    ndbi = np.where(np.isfinite(ndbi), ndbi, np.nan)

    return ndvi, ndbi


def classify_urban(ndbi, threshold=0.1):
    """
    Urban classification using NDBI threshold.
    """
    return np.where(ndbi > threshold, 1, 0).astype("uint8")


# =====================================================
# MAIN PROCESSING
# =====================================================

stats = []
processed_scenes = []

print("\nFinding Landsat scenes from 2015 to 2020...")
all_scenes = find_landsat_scenes()

print(f"Found {len(all_scenes)} scenes")

for scene in tqdm(all_scenes, desc="Processing Landsat scenes"):

    scene_files, missing_bands = find_scene_files(scene)

    if missing_bands:
        print(f"Missing bands for {scene['date'].strftime('%Y-%m-%d')}: {missing_bands}")
        continue

    try:
        band_data = {}

        for band_name, file_path in scene_files.items():
            arr, meta = clip_to_aoi(file_path, aoi)
            band_data[band_name] = arr

        red = band_data["red"]
        nir = band_data["nir"]
        swir = band_data["swir"]

        ndvi, ndbi = calc_indices(red, nir, swir)
        urban = classify_urban(ndbi, threshold=0.1)

        date_str = scene["date"].strftime("%Y%m%d")

        # Save NDVI
        meta_ndvi = meta.copy()
        meta_ndvi.update(dtype=rasterio.float32, count=1, nodata=np.nan)

        with rasterio.open(OUTPUT_DIR / f"NDVI_{date_str}.tif", "w", **meta_ndvi) as dst:
            dst.write(ndvi.astype("float32"), 1)

        # Save NDBI
        with rasterio.open(OUTPUT_DIR / f"NDBI_{date_str}.tif", "w", **meta_ndvi) as dst:
            dst.write(ndbi.astype("float32"), 1)

        # Save Urban map
        meta_urban = meta.copy()
        meta_urban.update(dtype=rasterio.uint8, count=1, nodata=0)

        with rasterio.open(OUTPUT_DIR / f"Urban_{date_str}.tif", "w", **meta_urban) as dst:
            dst.write(urban, 1)

        # Calculate urban area
        urban_pixels = np.sum(urban == 1)
        urban_area_km2 = urban_pixels * 900 / 1_000_000

        stats.append({
            "Date": scene["date"],
            "Year": scene["year"],
            "Month": scene["month"],
            "Urban_Pixels": urban_pixels,
            "Urban_Area_km2": urban_area_km2
        })

        processed_scenes.append(scene)

        print(f"Processed {scene['date'].strftime('%Y-%m-%d')} | Urban area: {urban_area_km2:.3f} km²")

    except Exception as e:
        print(f"Error processing {scene['date'].strftime('%Y-%m-%d')}: {e}")
        continue


# =====================================================
# SAVE RESULTS AND PLOTS
# =====================================================

if len(stats) == 0:
    print("\nNo scenes were processed. Check your paths and band filenames.")
else:
    df = pd.DataFrame(stats)
    df = df.sort_values("Date")

    csv_path = OUTPUT_DIR / "Urbanization_Stats_2015_2020.csv"
    df.to_csv(csv_path, index=False)

    print(f"\nStatistics saved: {csv_path}")
    print(df.head())

    # Monthly trend plot
    plt.figure(figsize=(15, 8))
    plt.plot(df["Date"], df["Urban_Area_km2"], marker="o", linewidth=1)
    plt.title("Urban Area Changes in Kurunegala District (2015–2020)")
    plt.xlabel("Date")
    plt.ylabel("Urban Area (km²)")
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "Urbanization_Trend_2015_2020.png", dpi=200)
    plt.show()

    # Yearly mean trend
    yearly_stats = df.groupby("Year")["Urban_Area_km2"].agg(["mean", "min", "max", "std"]).reset_index()

    plt.figure(figsize=(12, 6))
    plt.plot(yearly_stats["Year"], yearly_stats["mean"], marker="o", linewidth=2)
    plt.fill_between(
        yearly_stats["Year"],
        yearly_stats["min"],
        yearly_stats["max"],
        alpha=0.2,
        label="Min-Max Range"
    )
    plt.title("Yearly Urban Area Trend in Kurunegala District (2015–2020)")
    plt.xlabel("Year")
    plt.ylabel("Mean Urban Area (km²)")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(OUTPUT_DIR / "Yearly_Urbanization_Trend_2015_2020.png", dpi=200)
    plt.show()

    # Change map: first scene vs last scene
    if len(processed_scenes) >= 2:
        first_scene = processed_scenes[0]
        last_scene = processed_senes[-1] if False else processed_scenes[-1]

        first_date = first_scene["date"].strftime("%Y%m%d")
        last_date = last_scene["date"].strftime("%Y%m%d")

        urban_first_path = OUTPUT_DIR / f"Urban_{first_date}.tif"
        urban_last_path = OUTPUT_DIR / f"Urban_{last_date}.tif"

        with rasterio.open(urban_first_path) as src:
            urban_first = src.read(1)
            change_meta = src.meta.copy()

        with rasterio.open(urban_last_path) as src:
            urban_last = src.read(1)

        change = urban_last.astype("int8") - urban_first.astype("int8")

        change_meta.update(dtype=rasterio.int8, nodata=0)

        change_path = OUTPUT_DIR / f"UrbanChange_{first_date}_{last_date}.tif"

        with rasterio.open(change_path, "w", **change_meta) as dst:
            dst.write(change.astype("int8"), 1)

        print(f"Urban change map saved: {change_path}")

        plt.figure(figsize=(10, 8))
        plt.imshow(change, cmap="RdBu_r", vmin=-1, vmax=1)
        plt.title(
            f"Urban Change in Kurunegala\n{first_scene['date'].strftime('%Y-%m-%d')} to {last_scene['date'].strftime('%Y-%m-%d')}"
        )
        plt.colorbar(label="Urban Change: -1 Lost, 0 No Change, 1 New Urban")
        plt.axis("off")
        plt.tight_layout()
        plt.savefig(OUTPUT_DIR / "Urban_Change_Map_2015_2020.png", dpi=200)
        plt.show()

    print(f"\nUrbanization analysis completed.")
    print(f"Processed scenes: {len(stats)}")
    print(f"Results saved in: {OUTPUT_DIR}")
```
